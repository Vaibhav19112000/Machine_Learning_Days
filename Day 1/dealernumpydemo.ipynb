f.	Draw 2 bar chart one for each dealer, for each year total number of items sold in the year

```python
import matplotlib.pyplot as plt

x=[2001,2002,2003,2004,2005,2006,2007,2008,2009,2010]
y=dealer1.sum(axis=1)

plt.bar(x,y,label="sale",color="green")
plt.xticks(x);
plt.xlabel('year')
#this is for giving label to y axis
plt.ylabel('total sale')
plt.legend()
#plt.grid()
plt.title('year total number of items sold')
plt.show()
```

g.	Draw a pie chart one for each dealer, 

```python
import matplotlib.pyplot as plt

slices=dealer1.sum(axis=1)
years=[2001,2002,2003,2004,2005,2006,2007,2008,2009,2010]
cols=['g','c','m','r','k','y','pink','palegreen','r','b']

#to draw piechart in anticlockwise direction
#counterclock=False

#plt.pie(slices,labels=activities,colors=cols,startangle=90,counterclock=False)
plt.pie(slices,labels=years,colors=cols,shadow=True,explode=(0,0.2,0,0,0,0,0,0,0,0),startangle=90)

plt.title('year total number of items sold')
plt.show()
```
